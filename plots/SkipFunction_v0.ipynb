```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
noskip = pd.read_csv("data/noskip.csv")
skipall = pd.read_csv("data/skipall.csv")
skiphalf = pd.read_csv("data/skiphalf.csv")
skipthird = pd.read_csv("data/skipthird.csv")

threshold = noskip.iloc[:,0] #The thresholds are aligned for all datasets.
#Now get rates
rate_n = noskip.iloc[:,1]
rate_a = skipall.iloc[:,1]
rate_2 = skiphalf.iloc[:,1]
rate_3 = skipthird.iloc[:,1]

#Remove variables that won't be used again.
del noskip,skipall,skiphalf,skipthird
```

```python
plt.figure(figsize=(16,9))

plt.scatter(threshold,rate_n,label="No Skipping (skip = -1)")
plt.scatter(threshold,rate_a,label="Skip Everything (Skip = 1)")
plt.scatter(threshold,rate_2,label="Skip = 3")
plt.scatter(threshold,rate_3,label="Skip = 4")

plt.xlabel("Threshold",size=20)
plt.ylabel("Rate (Hz)",size=20)
plt.title("Trigger Rate vs Upper Threshold",size=30)
plt.rc('xtick',labelsize=16)
plt.rc('ytick',labelsize=16)

plt.legend()
plt.show()
```
